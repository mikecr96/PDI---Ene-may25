```python
"""
uint8 (0 - 255)
uint8 -> float32
SUCEDE LA MAGIA (hago las operaciones pertintentes)
float32 -> uint8
"""
```

```python
babuino_gray.max()
```

```python
babuino_gray_float32 = np.float64(babuino_gray.copy())
GAMMA = 9
babuino_gray_power = np.power(babuino_gray_float32, GAMMA)
babuino_gray_uint8 = np.uint8(babuino_gray_power)
plt.imshow(babuino_gray_uint8, cmap="gray")
plt.show()
```

```python
babuino_gray_float32 = np.float32(babuino_gray.copy())
babuino_gray_float32 = babuino_gray_float32 / 255.
# babuino_gray_float32.max()
GAMMA = 9
babuino_gray_power = np.power(babuino_gray_float32, GAMMA)
babuino_gray_uint8 = np.uint8(babuino_gray_power * 255) 
plt.imshow(babuino_gray_uint8, cmap="gray")
plt.show()
```
